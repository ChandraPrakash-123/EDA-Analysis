```python
data_filter=data[data['year']<2006]
data_filter.tail()
```

```python
df4=data_filter[['year','numcol','yieldpercol']]
df5=df4.groupby('year')[['numcol','yieldpercol']].sum()
df5
```

```python


# Create subplots
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Create bar plot in the first subplot
sns.barplot(data=df5, y='numcol', x=df5.index, hue='yieldpercol', ax=axs[0])
axs[0].set_title('year wise relation bitween numcol and yieldpercol',fontsize=14 )

# Create line plot in the second subplot
sns.lineplot(data=df5, y='numcol', x=df5.index, ax=axs[1])
axs[1].set_title('year wise relation bitween numcol and yieldpercol',fontsize=14 )

# Rotate x-axis labels
plt.xticks(rotation=90)

# Show the subplots
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))
ax2 = ax1.twinx()

# Plot the first line on the left y-axis:
sns.lineplot(x=df5.index, y='numcol', data=df5, ax=ax1)

# Plot the second line on the right y-axis:
sns.lineplot(x=df5.index, y='yieldpercol', data=df5, ax=ax2, color='orange')

# Set axis labels and title:
ax1.set_xlabel('Year')
ax1.set_ylabel('numcol')
ax2.set_ylabel('yieldpercol')
plt.title('Relationship between number of honey producing colony  and yield per colony', fontsize=14 )

plt.grid(linestyle='--')
plt.show()
```

## # conclusion:
- number of honey producing colony  decrease from 2002 to 2005 very rapidly .at a same time period yield of honey  per colony 
- decrease overall graph shows number of honey producing colony and yield reduce with time.

```python
grouping_prod=data.groupby('year')[['totalprod','prodvalue']].mean()
grouping_prod
```

```python
fig,axs=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
sns.lineplot(data=grouping_prod,y='totalprod',x='year',ax=axs[0],color="green")
axs[0].set_title('totalprod and year')
axs[0].set_xlabel('year ',fontsize=12)
axs[0].set_ylabel('totalprod',fontsize=12)
sns.lineplot(data=grouping_prod,y='prodvalue',x='year',ax=axs[1],color='blue')
axs[1].set_title('prodvalue and year',fontsize=14 )
axs[1].set_xlabel('year',fontsize=12)
axs[1].set_ylabel('prodvalue',fontsize=12)
plt.show()
```
